```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/sms-spam-collection-dataset")

print("Path to dataset files:", path)
```

```python
import os
import pandas as pd
print(os.listdir(path))

df=pd.read_csv(f'{path}/spam.csv', encoding='latin-1')
```

```python
df=df[['v1', 'v2']]
df.columns=['label', 'message']
```

```python
df['label']=df['label'].map({'ham':0, 'spam':1})
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
x=vectorizer.fit_transform(df['message'])
y=df['label']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
```

```python
print(model.score(x_test, y_test))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
def predict_message(msg):
    msg_x=vectorizer.transform([msg])
    prediction=model.predict(msg_x)
    return "Spam" if prediction[0]==1 else "Not Spam"
```

```python
print(predict_message("Congratulations! You won a free iPhone"))
print(predict_message("Hey, are we meeting today?"))
```

```python
import pickle
with open("spam_det_model.pkl", "wb") as f:
    pickle.dump(model, f)
with open("vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)
```
